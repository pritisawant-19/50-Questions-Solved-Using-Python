#### 9. Reverse a number using a loop.

```python
def reverse_num(n):
    rev = 0
    while n:
        rev = rev*10 + n%10
        n //= 10
    return rev

print(reverse_num(12345))
```

#### 15. Count vowels in a string

```python
s = "beautiful day"
vowels = "aeiouAEIOU"
count = 0

for ch in s:
    if ch in vowels:
        count += 1

print("Vowel count:", count)
```

#### 16. Count consonants in a string

```python
s = "Good Morning"
vowels = "aeiouAEIOU"
count = 0

for ch in s:
    if ch.isalpha() and ch not in vowels:
        count += 1

print("Consonant count:", count)
```

#### 22. Check if a number is prime

```python
num = 29   

if num < 2:
    print("Not Prime")
else:
    is_prime = True
    for i in range(2, num):
        if num % i == 0:
            is_prime = False
            break

    if is_prime:
        print("Prime")
    else:
        print("Not Prime")
```

#### 23. Print all prime numbers from 1 to N.

```python
N = 50   

for num in range(2, N + 1):
    is_prime = True
    for i in range(2, num):
        if num % i == 0:
            is_prime = False
            break
    if is_prime:
        print(num, end=" ")
```
